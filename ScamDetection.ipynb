```python
# Function to detect scams based on keywords
def is_scam(transcript):
    scam_keywords = ["win", "free", "prize", "urgent", "call now", "limited time","offer", "otp", "kyc","link"]
    for keyword in scam_keywords:
        if re.search(r'\b' + keyword + r'\b', transcript, re.IGNORECASE):
            return True
    return False
```
